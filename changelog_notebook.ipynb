```python
# fetch the function we will use from utils.py
from utils import fetch_trending_projects_from_changelog, get_ghdata_from_fullname
import pandas as pd
```

```python
# let's fetch data for June 1st, 2018 and the 10 days that follow
from datetime import datetime

analysis_date = datetime(2018,6,1)

df = fetch_trending_projects_from_changelog(
    start_date=analysis_date,
    duration_days=10
)
```

```python
# let's view the data
df.head().T
```

```python
# fetch single data row based on project
data_row = get_ghdata_from_fullname("facebook/Sonar")
```

```python
data_row
```

```python
# fetch data for all projects found in changelog

# keep track of fetched projects
project_fetched = set()

# list to store data rows
gh_data = []

# iterate of Pandas dataframe
for idx, row in df.iterrows():
    full_name = row.get("full_name")
    if full_name in project_fetched:
        print("skipping {} ... already done!".format(full_name))
        continue
        
    gh_data.append(get_ghdata_from_fullname(full_name))

    # project data fetched
    project_fetched.add(full_name)

```

```python
# not all data was fetched, but what we got so far will do

# make dataframe from valid GitHub data
gh_df = pd.DataFrame([x for x in gh_data if x is not None])
```

```python
# view data
gh_df.head().T
```

```python
# lets merge GitHub data with ChangeLog data
merged_df = df.merge(gh_df, left_on="full_name", right_on="gh_full_name", how="inner")
```

```python
# view merged dataframe
merged_df.head().T
```

```python
# to get only the last observation for each project
last_df = merged_df.groupby("full_name").last()
```

```python
# view df
last_df.head().T
```

```python
# store data as csv file, exclude dataframe index
last_df.to_csv("github_and_changelog.csv", index=False)
```
